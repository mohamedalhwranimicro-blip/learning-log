```python
import requests

def search_images(term, max_images=1):
    url = f"https://www.bing.com/images/search?q={term}&form=HDRSC2"
    headers = {"User-Agent": "Mozilla/5.0"}
    r = requests.get(url, headers=headers)
    return [i.split('murl&quot;:&quot;')[1].split('&quot;')[0] 
            for i in r.text.split('murl&quot;:&quot;')[1:max_images+1]]
```

```python
urls = search_images("Laptop photos", max_images=3)
if urls:
    urls[0]
else:
    print("No images found")
```
